```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('test_data.csv')
# print(data)

def razans_implementation(points: pd.DataFrame,w=0,b=0,lr=0.001,epochs=100):
    '''HOME MADE FUNCTION'''
    n=len(points)
    loss_list=[]
    for epoch in range(epochs):
        for i in range(n):
            x=points.iloc[i].x
            y=points.iloc[i].y

            y_pred = b+w*x
            w_slope = (y_pred-y)*2*x
            b_slope = 2*(y_pred-y)
            loss = (y-y_pred)**2
            w = w-lr*w_slope
            b = b-lr*b_slope

        # print(f"Epoch Number: {epoch} || Loss: {loss}")
        loss_list.append(loss)
    print(len(loss_list))
    plt.scatter(range(len(loss_list)),loss_list)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    print("weight: ",w,"bias: ", b)
    return w,b
w,b=razans_implementation(data)

```
